```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
numIterations = 5000

# Weights in the network with initial values
w1 = 0.1
w2 = 0.2
w3 = 0.3
w4 = 0.4
w5 = 0.5
w6 = 0.6
w7 = 0.7
w8 = 0.8
w9 = 0.9
w10 = 0.1
weights = [w1, w2, w3, w4, w5, w6, w7, w8, w9, w10]
```

```python
# Biases in the network with initial values
b1 = 0.5
b2 = 0.5
biases = [b1, b2]
```

```python
# Original input from user
x1 = 1
x2 = 4
x3 = 5
xList = [x1,x2,x3]
```

```python
# Target putput expected
t1 = 0.1
t2 = 0.05
targets = [t1, t2]
```

```python
# Hyperparameter
learningRate = 0.01
```

```python
# Define nonlinear activation function with sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Calculate outputs by forward propagation through the network
def forwardPropagation(xList, weights, biases):
    # value before activation to the h1 on the hidden layer
    zh1 = weights[0] * xList[0] + weights[2] * xList[1] + weights[4] * xList[2] + biases[0]
    
    # value before activation to the h2 on the hidden layer
    zh2 = weights[1] * xList[0] + weights[3] * xList[1] + weights[5] * xList[2] + biases[0]
    
    # value after activation
    h1 = sigmoid(zh1)
    h2 = sigmoid(zh2)
    
    # value after activation to the o1 on the output layer
    zo1 = weights[6] * h1 + weights[8] * h2 + biases[1]
    
    # value after activation to the o2 on the output layer
    zo2 = weights[7] * h1 + weights[9] * h2 + biases[1]
    
    # value after activation
    o1 = sigmoid(zo1)
    o2 = sigmoid(zo2)
    
    return h1, h2, o1, o2
```

```python
# Define loss function
def error(outputs, targets):
    return 1 / 2 * (np.power(outputs[0] - targets[0], 2) + np.power(outputs[1] - targets[1], 2))
```

```python
def CalculatePartialDerivative():
    # Error partial derivative with resprct to w7
    dE_do1 = o1 - t1
    do1_dzo1 = o1 * (1 - o1)
    dzo1_dw7 = h1
    dE_dw7 = dE_do1 * do1_dzo1 * dzo1_dw7
    
    # Error partial derivative with resprct to w8
    dE_do2 = o2 - t2
    do2_dzo2 = o2 * (1 - o2)
    dzo2_dw8 = h1
    dE_dw8 = dE_do2 * do2_dzo2 * dzo2_dw8
    
    # Error partial derivative with respect to w9
    dzo1_dw9 = h2
    dE_dw9 = dE_do1 * do1_dzo1 * dzo1_dw9
    
    # Error partial derivative with respect to w10
    dzo2_dw10 = h2
    dE_dw10 = dE_do2 * do2_dzo2 * dzo2_dw10
    
    # Error partial derivative with respect to b2
    dzo1_db2 = 1
    dzo2_db2 = 1
    dE_db2 = dE_do1 * do1_dzo1 * dzo1_db2 + dE_do2 * do2_dzo2 * dzo2_db2
    
    # Error partial derivative with respect to h1
    dzo1_dh1 = w7
    dzo2_dh1 = w8
    dE_dh1 = dE_do1 * do1_dzo1 * dzo1_dh1 + dE_do2 * do2_dzo2 * dzo2_dh1
    
    # Error partial derivative with respect to w1
    dh1_dzh1 = h1 * (1 - h1)
    dzh1_dw1 = x1
    dE_dw1 = dE_dh1 * dh1_dzh1 * dzh1_dw1
    
    # Error partial derivative with respect to w3
    dzh1_dw3 = x2
    dE_dw3 = dE_dh1 * dh1_dzh1 * dzh1_dw3
    
    # Error partial derivative with respect to w5
    dzh1_dw5 = x3
    dE_dw5 = dE_dh1 * dh1_dzh1 * dzh1_dw5
    
    # Error partial derivative with respect to h2
    dzo1_dh2 = w9
    dzo2_dh2 = w10
    dE_dh2 = dE_do1 * do1_dzo1 * dzo1_dh2 + dE_do2 * do2_dzo2 * dzo2_dh2
    
    # Error partial derivative with respect to w2
    dh2_dzh2 = h2 * (1 - h2)
    dzh2_dw2 = x1
    dE_dw2 = dE_dh2 * dh2_dzh2 * dzh2_dw2
    
    # Error partial derivative with respect to w4
    dzh2_dw4 = x2
    dE_dw4 = dE_dh2 * dh2_dzh2 * dzh2_dw4
    
    # Error partial derivative with respect to w6
    dzh2_dw6 = x3
    dE_dw6 = dE_dh2 * dh2_dzh2 * dzh2_dw6
    
    # Error partial derivative with respct to b1
    dzh1_db1 = 1
    dzh2_db1 = 1
    dE_db1 = ((dE_do1 * do1_dzo1 * dzo1_dh1 * dh1_dzh1 * dzh1_db1)
             + (dE_do2 * do2_dzo2 * dzo2_dh2 * dh2_dzh2 * dzh2_db1))
    
    return (dE_dw1, dE_dw2, dE_dw3, dE_dw4, dE_dw5, dE_dw6,
            dE_dw7, dE_dw8, dE_dw9, dE_dw10, dE_db1, dE_db2)
```

```python
# Historical errors (cost)
errors = []

for i in range(numIterations):
    # Execute forward propagation one time
    h1, h2, o1, o2 = forwardPropagation(xList, weights, biases)
    
    # Calculate SSE (Sum of Square Error) error accuracy
    sse = error([o1, o2], targets)
    
    # Store historical errors for final plotting
    errors.append(sse)
    
    print(f'Running {i + 1} of {numIterations}')
    print(f'w1: {w1}')
    print(f'w2: {w2}')
    print(f'w3: {w3}')
    print(f'w4: {w4}')
    print(f'w5: {w5}')
    print(f'w6: {w6}')
    print(f'w7: {w7}')
    print(f'w8: {w8}')
    print(f'w9: {w9}')
    print(f'w10: {w10}')
    
    print(f'b1: {b1}')
    print(f'b2: {b2}')
    
    print(f'o1: {o1}')
    print(f'o2: {o2}')
    print(f't1: {t1}')
    print(f't2: {t2}')
    print(f'error: {sse}')
    print('\n')
    
    (dE_dw1, dE_dw2, dE_dw3, dE_dw4, dE_dw5, dE_dw6,
     dE_dw7, dE_dw8, dE_dw9, dE_dw10, dE_db1, dE_db2) = CalculatePartialDerivative()
    
    w1 -= (learningRate * dE_dw1)
    w2 -= (learningRate * dE_dw2)
    w3 -= (learningRate * dE_dw3)
    w4 -= (learningRate * dE_dw4)
    w5 -= (learningRate * dE_dw5)
    w6 -= (learningRate * dE_dw6)
    w7 -= (learningRate * dE_dw7)
    w8 -= (learningRate * dE_dw8)
    w9 -= (learningRate * dE_dw9)
    w10 -= (learningRate * dE_dw10)
    
    b1 -= (learningRate * dE_db1)
    b2 -= (learningRate * dE_db2)
    
    weights = [w1, w2, w3, w4, w5, w6, w7, w8, w9, w10]
    biases = [b1, b2]
```

```python
import numpy as np
import pandas as pd
import torch as torch
import torch.nn.functional as F
import crypten
torch.set_printoptions(precision=16)
torch.set_default_tensor_type(torch.DoubleTensor)
crypten.init()
```

```python
class network():
    def __init__(self, in_size, hid_size, out_size, rate):
        assert len(hid_size) > 0, 'a non-empty vector is expected'
        self.in_size = in_size
        self.hid_size = hid_size
        self.out_size = out_size
        self.rate = rate
        self.cons_net()
        print(f'network initialization finished')
        
    def cons_net(self):
        self.biases = torch.from_numpy(np.random.rand(len(self.hid_size) + 1))
        self.weights = []
        
        w_in = torch.from_numpy(np.random.rand(self.in_size, self.hid_size[0]))
        self.weights.append(w_in)
        
        if len(self.hid_size) > 1:
            for index in range(len(self.hid_size) - 1):
                w = torch.from_numpy(np.random.rand(self.hid_size[index], self.hid_size[index + 1]))
                self.weights.append(w)
        
        w_out = torch.from_numpy(np.random.rand(self.hid_size[-1], self.out_size))
        self.weights.append(w_out)
        
    def forward(self, x_enc):
        self.zh = []
        self.h = []
        
        zh1 = x_enc.matmul(self.weights[0]) + self.biases[0]
        self.zh.append(zh1)
        h1 = zh1.sigmoid()
        self.h.append(h1)
        
        index = 1
        
        for weight in self.weights[1:-1]:
            zh_temp = self.h[-1].matmul(weight) + self.biases[index]
            self.zh.append(zh_temp)
            h_temp = zh_temp.sigmoid()
            self.h.append(h_temp)
            index = index + 1
        
        self.zo = self.h[-1].matmul(self.weights[-1]) + self.biases[-1]
        o = self.zo.sigmoid()
        return o
    
    def backPropagation(self, o, t, x_enc):
        dE_do = o - crypten.cryptensor(t)
        do_dzo = o * (1 - o)

        dE_dzo = (dE_do * do_dzo).get_plain_text()

        dzo_dw = self.h[-1].get_plain_text()

        dE_dw = dzo_dw.T.matmul(dE_dzo)

        self.weights[-1] = self.weights[-1] - dE_dw  * self.rate

        dE_dzh = dE_dzo

        for i in range(len(self.h) - 1, 0, -1):
            dzhiii_dhii = self.weights[i + 1]
            dhii_dzhii = self.h[i].get_plain_text().T.matmul(1 - self.h[i].get_plain_text())
    
            dE_dhii = dzhiii_dhii.matmul(dE_dzh.T.double())
            dE_dzh = dE_dhii.T.matmul(dhii_dzhii.double())
    
            dzhi_dwi = self.h[i - 1].get_plain_text()
    
            dE_dwi = dzhi_dwi.T.double().matmul(dE_dzh)
    
            self.weights[i] = self.weights[i] - dE_dwi * self.rate
    
        dzh3_dh2 = self.weights[1]
        dh2_dzh2 = self.h[0].get_plain_text().T.matmul(1 - self.h[0].get_plain_text())

        dE_dh2 = dzh3_dh2.matmul(dE_dzh.T.double())
        dE_dzh = dE_dh2.T.matmul(dh2_dzh2.double())

        dzh1_dw1 = x_enc.get_plain_text()

        dE_dw1 = dzh1_dw1.T.double().matmul(dE_dzh)

        self.weights[0] = self.weights[0] - dE_dw1 * self.rate
```

```python
def crossEntropy(o, t):
    result = 0.
    for i in range(len(o)):
        result += pow(o[i] - t[i], 2)
    return result / len(o)

crossEntropy([1.,0.,0.], [0.5, 0.,0.])
```

```python
nn = network(4, [5,6,7,5,6], 3, 0.1)
x = [[3.,2.,3.,1.]]
x_enc = crypten.cryptensor(x)
t = torch.tensor([[1.,0.,0.]])

for i in range(100):
    o = nn.forward(x_enc)
    loss = crossEntropy(o.get_plain_text()[0], t[0])
    print(f'loss of epoch[{i + 1}/100] is: {loss}')
    nn.backPropagation(o, t, x_enc)
```
